```python
x=df.drop(["BasePay"],axis=1)
y=df["BasePay"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
ct=ColumnTransformer(transformers=[
    ("encoding",OrdinalEncoder(),list(x.select_dtypes("object").columns))
],remainder="passthrough")
```

```python
xtrain=ct.fit_transform(xtrain)
xtest=ct.fit_transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,accuracy_score
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
print(lr.score(xtrain,ytrain))
r2_score(ytest,ypred)
```

```python
def mymodel(modell):
    for model in modell:
        print("="*100)
        print("Algo Name         : ",model)
        model.fit(xtrain,ytrain)
        ypred=model.predict(xtest)
        print("training score is : ",model.score(xtrain,ytrain))
        print("Testing_score     : ",r2_score(ytest,ypred))
        mse = mean_squared_error(ytest, ypred)
        mae = mean_absolute_error(ytest, ypred)
        print("MSE               : ",mse)
        print("MSA               : ",mae)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
elsticnet=ElasticNet()
svr=SVR()
random=RandomForestRegressor(max_depth=4,random_state=45)
gradient=GradientBoostingRegressor(max_depth=2,random_state=42)
KNNR=KNeighborsRegressor(n_neighbors=4)
dt=DecisionTreeRegressor(max_depth=4,random_state=42)
lasso=Lasso()
ridge=Ridge()
lrr=LinearRegression()

model=[lrr,lasso,ridge,dt,svr,elsticnet,random,gradient,KNNR]

mymodel(model)
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import Lasso, Ridge, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor

elasticnet = ElasticNet()
svr = SVR()
random = RandomForestRegressor(max_depth=4)
gradient = GradientBoostingRegressor(max_depth=2)
KNNR = KNeighborsRegressor(n_neighbors=7)
dt = DecisionTreeRegressor(max_depth=3)
lasso = Lasso()
ridge = Ridge()
lrr = LinearRegression()

model = [
            ('lrr', lrr), 
         ('lasso', lasso), 
         ('ridge', ridge), 
         ('dt', dt), 
#          ('svr', svr), 
         ('elasticnet', elasticnet), 
         ('random', random), 
         ('gradient', gradient), 
#          ('KNNR', KNNR)
        ]
vt = VotingRegressor(estimators=model)

vt.fit(xtrain, ytrain)
ypred = vt.predict(xtest)

print("Training score is:", vt.score(xtrain, ytrain))
print("Testing score:", vt.score(xtest, ytest))
```
